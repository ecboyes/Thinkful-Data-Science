```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
admissions = pd.read_csv('Admission_Predict.csv',encoding = 'ISO-8859-1')

df = pd.DataFrame(admissions)
print(df)
```

```python
ax = sb.distplot(admissions["GRE Score"], kde=True, rug = True, color="blue")
#ax = sb.hist(x='GRE Score',y='CGPA', data=admissions)  
plt.title('Plot 1: GRE Scores for 399 Students')
sb.despine(offset=10, trim=True)
ax.set(xlabel='GRE Score (out of 340)')

plt.show()
```

```python
ax1 = sb.distplot(admissions["University Rating"], kde=True, color="pink", bins=np.linspace(1,5,5))
#ax1 = sb.distplot(data.column, hist_kws=dict(bins=np.linspace(1,5,5)))
#ax = sb.barplot(x='University Rating', y='Serial No.', data=admissions)  
plt.title('Plot 2: University Rating (out of 5)')
sb.despine(offset=10, trim=True)
ax1.set(xlabel='Rating')

plt.show()
```

```python
ax3 = sb.distplot(admissions["TOEFL Score"], kde=True, color="orange")
#ax = sb.hist(x='GRE Score',y='CGPA', data=admissions)  
plt.title('Plot 3: TOEFL Scores for 399 Students')
sb.despine(offset=10, trim=True)
ax3.set(xlabel='Score')

plt.show()
```

```python
ax4 = sb.distplot(admissions["LOR"], kde=True, color="green")
#ax = sb.hist(x='GRE Score',y='CGPA', data=admissions)  
plt.title('Plot 4: Letter of Recommendation Strength')
sb.despine(offset=10, trim=True)
ax4.set(xlabel='Strength (out of 5)')

plt.show()
```

```python
# Two continuous variables three different ways: College GPA vs. GRE Score

plt.figure(figsize=(9,6))
ax = sb.lineplot(x="GRE Score", y="CGPA", data=df, palette="turquoise")
ax.set_title("College GPA vs. GRE Score", fontsize=15)

```

```python
plt.figure(figsize=(9,6))
ax = sb.scatterplot(x="GRE Score", y="CGPA", data=df, color='violet')
ax.set_title("College GPA vs. GRE Score", fontsize=15)
```

```python
ax = sb.relplot(x="GRE Score", y="CGPA", hue='University Rating', data=df)
```

------------

```python
# One continuous variable and one categorical variable six different ways: University Rating vs. GRE Score

## Swarmplot

plt.figure(figsize=(9,6))
ax = sb.swarmplot(x="University Rating", y="GRE Score", data=df, color='goldenrod')
```

```python
# Catplot

ax = sb.catplot(x="University Rating", y="GRE Score", data=df, color='chartreuse')
```

```python
# Boxplot

plt.figure(figsize=(9,6))
ax = sb.boxplot(x="LOR", y="GRE Score", hue="University Rating", data=df, palette="Set2")
ax.set_title("Letter of Recommendation Strength vs. GRE Score", fontsize=15)
```

```python
# Pointplot

plt.figure(figsize=(9,6))
ax = sb.pointplot(x="University Rating", y="GRE Score", data=df, color='teal')
ax.set_title("University Rating vs. GRE Score", fontsize=15)
```

```python
# Violinplot

plt.figure(figsize=(9,6))
ax = sb.violinplot(x="University Rating", y="GRE Score", data=df, color='pink')
ax.set_title("University Rating vs. GRE Score", fontsize=15)
```

```python
# Countplot

plt.figure(figsize=(9,6))
ax = sb.countplot(hue="GRE Score", x="University Rating", data=df, color='red')
ax.set_title("University Rating vs. GRE Score", fontsize=15)
```
