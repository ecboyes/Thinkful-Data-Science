```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
weather_df = pd.read_sql_query('select * from weatherinszeged',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
```

```python
# Y is the target variable
Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set
X = weather_df[['humidity','windspeed']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The R squared and adjusted R squared values are both 0.288, which is a bit low, but still has some predictive power.

```python
weather_df['humidity_windspeed_interaction'] = weather_df.humidity * weather_df.windspeed

# Y is the target variable
Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set
X = weather_df[['humidity','windspeed', 'humidity_windspeed_interaction']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The R squared of this model is 0.341, which is an improvement upon the last model which had an R squared value of 0.288. This is still a bit low, but it is definitely an improvement over the last model.

```python
# Y is the target variable
Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set
X = weather_df[['humidity','windspeed', 'visibility']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The (adjusted) R squared of this model (0.303) is higher than the R squared of the first model (0.288), but lower than the R squared of the second model (0.341). The best model of the three is therefore the second model, which includes the interaction of the windspeed and humidity.

The AIC and BIC of the three models are:
1. 340,900
2. 333,400
3. 338,800

According to these numbers alone, the second model would be the best of the three. This prediction tracks with the R squared comparison of the three models.
