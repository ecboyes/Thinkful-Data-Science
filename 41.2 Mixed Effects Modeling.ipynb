```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Creating data for ten participants, with ten datapoints each.
df_combined=pd.DataFrame({"x":list(range(1,11))*10,
                "y":list(range(1,21,2))*10,
                "id":np.repeat(range(10),10),
                "time":list(range(10))*10})

#Differentiating the data from the different participant ids.
for i in range(10):
    df_combined.loc[df_combined['id']==i,"y"]=(df_combined.loc[df_combined['id']==i,"y"]+i*3) + (df_combined.loc[df_combined['id']==i,"y"]*i)


# Plotting the data: Same slope, different intercepts.
sns.lmplot(x="x", y="y", hue="id", data=df_combined)
plt.show()
```
